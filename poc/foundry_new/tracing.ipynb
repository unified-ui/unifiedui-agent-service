- **/openai/conversations/{id}/items?api-version=2025-11-15-preview**

```python
# https://learn.microsoft.com/de-de/azure/ai-foundry/quickstarts/get-started-code?view=foundry&tabs=rest
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
FOUNDRY_ACCESS_TOKEN = os.getenv("FOUNDRY_ACCCESS_TOKEN")
FOUNDRY_CONVERSATION_ID = "conv_c671246113e78fe700QvxJVpUm1OlmSlm4N81oS4QPxhQ2Og1b"
```

```python
import requests
import json

ep = f"https://engo-foundry.services.ai.azure.com/api/projects/proj-default-2/openai/conversations/{FOUNDRY_CONVERSATION_ID}/items?api-version=2025-11-15-preview"

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {FOUNDRY_ACCESS_TOKEN}"
}

response = requests.get(ep, headers=headers)
tracing_response = response.json()

filepath = f"resp_tracing_{FOUNDRY_CONVERSATION_ID}.json"
with open(filepath, "w") as f:
    f.write(json.dumps(tracing_response, indent=2))
tracing_response
```
